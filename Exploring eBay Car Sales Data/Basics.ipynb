```python
# convert price column to numeric datatype

autos["price"] = autos["price"].str.replace("$","").str.replace(",","")
autos["price"]= autos["price"].astype(int)
```

```python
#convert odometer to numeric datatype

autos["odometer"] = autos["odometer"].str.replace("km","").str.replace(",","")
autos["odometer"] = autos["odometer"].astype(int)
```

```python
auto_price = autos["brand"].value_counts(normalize= True)
brand_select = auto_price[auto_price > 0.025].index

price_dict= {}

for a in brand_select:
    brand_rows = autos[autos["brand"] == a]
    price_mean = brand_rows["price"].mean()
    price_dict[a] = int(price_mean)

price_dict
```

```python
mileage_dict = {}

for i in brand_select:
    brand_rows = autos[autos["brand"] == a]
    mileage_mean = brand_rows["odometer_km"].mean()
    mileage_dict[i] = int(mileage_mean)
    
mileage_dict
```

```python
df = pd.DataFrame (pd.Series(price_dict), columns = ['mean_price'])
df
```

```python
mileage_series = pd.Series(mileage_dict)
df["mileage_mean"] = mileage_series
df
```
